```python
# Create the grapth
import json
import matplotlib.pyplot as plt

JSON_folder_name = 'organized_data/20240723_run_3_gaus/'
name_models_choise = [
    'Clements_Arct', 'ClementsBell_Arct',
    'Fldzhyan_Arct', 'FldzhyanBell_Arct',
    'NEUROPULS_Arct', 'NEUROPULSCrossingSide_Arct',
    'NEUROPULSBell_Arct', 'NEUROPULSBell2_Arct',
    'NEUROPULSBellCrossingSide_Arct', 'NEUROPULSBellCrossingSide2_Arct',]

def filter_list_of_dicts(dict_list, PCILMU, PCILSIGMA, IMBMU):
    filtered_list = []
    for dictionary in dict_list:
        if list(dictionary.values())[0] == PCILMU and list(dictionary.values())[1] == PCILSIGMA and list(dictionary.values())[4] == IMBMU:
            filtered_list.append(dictionary)
    return filtered_list

average = []
std_dev = []
fig = plt.figure(figsize=(10, 6))
# Download the file
for arct in name_models_choise:
    file_JSON = "20240723_traking_"+arct+"_N8.json"
    with open(JSON_folder_name+file_JSON, 'r') as file:
        data = json.load(file)
    simulations = data["simulations"]
    filtered_list_disct = filter_list_of_dicts(dict_list=simulations, PCILMU=0.0, PCILSIGMA=0.0, IMBMU=0.0)
    average.extend([sim["average"] for sim in filtered_list_disct])
    std_dev.extend([sim["std_dev"] for sim in filtered_list_disct])

plt.errorbar(name_models_choise, average, yerr=std_dev, fmt='o')
plt.xlabel('Phase Changers insertion loss [dB]')
plt.ylabel('Fidelity')
fig.autofmt_xdate()
plt.xlabel("Architectures")
fig.subplots_adjust(bottom=0.3)
plt.grid(axis='y')
plt.show()
```
